```python
import numpy as np
import matplotlib.pyplot as plt

def posisi(t: np.ndarray) -> np.ndarray:
  return t ** 3 - 6 * t ** 2 + 9 * t

def turunan(t: np.ndarray) -> np.ndarray:
  return 3 * t ** 2 - 12 * t + 9

def hitung(t_value: np.ndarray, s_value: np.ndarray, v_value: np.ndarray, waktu_evaluasi: float):
  fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

  ax[0].plot(t_value, s_value, label="posisi $s(t)$", color='blue')
  ax[0].set_title("grafik antara posisi benda ke waktu")
  ax[0].set_ylabel("posisi dalam bentuk (meter)")
  ax[0].grid(True)
  ax[0].legend()

  ax[1].plot(t_value, v_value, label="kecepatan $v(t)$", color='green')
  ax[1].axvline(x=waktu_evaluasi, color='red', linestyle='--', label=f"evaluasi waktu t={waktu_evaluasi}")
  ax[1].set_title("grafik dari kecepatan benda ke waktu")
  ax[1].set_xlabel("waktu (detik)")
  ax[1].set_ylabel("kecepatan (meter / detik)")
  ax[1].grid(True)
  ax[1].legend()

  plt.tight_layout()
  plt.show()
```

```python
waktu_start = 0.0
waktu_akhir = 4.0
point_angka = 400

t_value = np.linspace(waktu_start, waktu_akhir, point_angka)
s_value = posisi(t_value)
v_value = turunan(t_value)

waktu_evaluasi = 2.0
waktu_tururnan_evaluasi = turunan(waktu_evaluasi)

hitung(t_value, s_value, v_value, waktu_evaluasi)
```
